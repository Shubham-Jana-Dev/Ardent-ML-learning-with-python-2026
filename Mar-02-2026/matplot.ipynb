```python
# Pyplot API
# Univariate - Numarical
data = {'salary': [1600,19000,25000,3000,35000,40000,50000,40000,90000,120000,150000]}
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
# line plot:
plt.plot(df['salary'],color = 'red', marker = '*', linestyle = ":", linewidth = '3')
plt.grid()
plt.show()
```

```python
# Histogram
plt.grid()
plt.hist(df['salary'],bins = 5)
plt.show()
```

```python
plt.boxplot(df['salary'])
plt.show()
```
