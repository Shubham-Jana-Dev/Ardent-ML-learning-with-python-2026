```python
# Pyplot API
# Univariate - Numarical
data = {'salary': [1600,19000,25000,3000,35000,40000,50000,40000,90000,120000,150000]}
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
# line plot:
plt.plot(df['salary'],color = 'red', marker = '*', linestyle = ":", linewidth = '3')
plt.grid()
plt.show()
```

```python
# Histogram
plt.grid()
plt.hist(df['salary'],bins = 5)
plt.show()
```

```python
plt.boxplot(df['salary'])
plt.show()
```

```python
df.loc[20]=[0]
```

```python
plt.boxplot(df['salary'])
plt.show()
```

```python
# Pie chart:
```

```python
count = df["dept"].value_counts()
count
```

```python
plt.pie(count)
plt.show()
```

```python
plt.pie(count, labels = count.index)
plt.show()
```

```python
plt.pie(count, labels = count.index, autopct = '%1.1f')
plt.show()   
```

```python
plt.pie(count, labels = count.index, autopct = "%1.2f", explode = [0, 0.1, 0])
plt.axis('equal')
plt.show()
```

```python
plt.bar(count.index, count)
plt.show()     
```
